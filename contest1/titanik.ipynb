```python
useless_features = ['Name', 'Ticket','Cabin']

helpfull_train = train.drop(columns=useless_features)
helpfull_test = test.drop(columns=useless_features)
helpfull_train
```

```python
def impute_data(df_copy):
  #Заполняем пропущенные категориальные фичи модой
  categorial_features = list(df_copy.select_dtypes(include=['object']).columns.values)
  for feature in categorial_features:
    if (df_copy[feature].isnull().values.any()):
      df_copy[feature].fillna(df_copy[feature].mode()[0], inplace=True)

  numerical_features = list(df_copy.select_dtypes(include=['int64', 'float64']).columns.values)
  for feature in numerical_features:
    if (df_copy[feature].isnull().values.any()):
      df_copy[feature].fillna(df_copy[feature].median(), inplace=True)

  return df_copy
```

```python
imputed = impute_data(helpfull_train)
imputed.info()
imputed['Survived'] = imputed['Survived'].astype(str)
```

```python
sns.catplot(x='Sex', data=imputed, hue='Survived', kind='count')
```
